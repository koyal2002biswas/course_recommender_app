```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print('Dependencies Imported')
```

```python
data=pd.read_csv('Coursera.csv')
```

```python
dup=data
dup
```

```python
dup.drop(['University'], axis=1, inplace =True)
dup
```

```python
dup.drop_duplicates(subset='Course Name', keep='first', inplace =True)
dup
```

```python
dup = dup[['Course Name','Difficulty Level','Course Description','Skills', 'Course URL']]
```

```python
dup
```

```python
dup['Course Name'] = dup['Course Name'].str.replace(' ',',')
dup['Course Name'] = dup['Course Name'].str.replace(',,',',')
dup['Course Name'] = dup['Course Name'].str.replace(':','')
dup['Course Description'] = dup['Course Description'].str.replace(' ',',')
dup['Course Description'] = dup['Course Description'].str.replace(',,',',')
dup['Course Description'] = dup['Course Description'].str.replace('_','')
dup['Course Description'] = dup['Course Description'].str.replace(':','')
dup['Course Description'] = dup['Course Description'].str.replace('(','')
dup['Course Description'] = dup['Course Description'].str.replace(')','')

#removing paranthesis from skills columns 
dup['Skills'] = dup['Skills'].str.replace('(','')
dup['Skills'] = dup['Skills'].str.replace(')','')
```

```python
dup['tags'] = dup['Course Name'] + dup['Difficulty Level'] + dup['Course Description'] + dup['Skills']
new_df = dup[['Course Name','tags','Difficulty Level' ,'Course URL']]
new_df.head(3)
```

```python
new_df['tags'] = dup['tags'].str.replace(',',' ')
new_df['Course Name'] = dup['Course Name'].str.replace(',',' ')
new_df.rename(columns = {'Course Name':'course_name'}, inplace = True)
```

```python
new_df.rename(columns = {'Course URL':'course_url'}, inplace = True)
```

```python
print(new_df.shape)
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()

import nltk #for stemming process
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(course):
    course_index = new_df[new_df['course_name'] == course].index[0]
#     url_index = new_df[new_df['course_url'] == course].index[4]
    
    distances = similarity[course_index]
    course_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:7]
    
    for i in course_list:
        print(new_df.iloc[i[0]].course_name,new_df.iloc[i[0]].course_url )
#     print(course_index)
```

```python
recommend('Business Strategy Business Model Canvas Analysis with Miro')
```
